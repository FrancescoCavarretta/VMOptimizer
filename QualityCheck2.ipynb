```python
import matplotlib.pyplot as plt
import numpy as np

notes = [ '' ] * 33

notes[0] = 'Low AP peak, below 0'
notes[6] = 'T-Ca2+ still too active'
notes[13] = 'T-Ca2+ still too active, not great AHP shape'
notes[14] = 'AHP below baseline'
notes[20] = 'Weird AHP shape'
notes[27] = 'Weird AHP shape'
notes[31] = 'Weird AHP shape'

for i in range(33):
    trace = np.load('trace_chk1_%d.npy' % i, allow_pickle=True).tolist()
    print('%s %d %d | %d' % (trace['key'] + (i, )))
    plt.figure(figsize=(15, 5))
    for j, (lbl, tr) in enumerate(sorted(trace['responses'].items())):
        plt.subplot(1, 6, j+1)
        plt.title(lbl)
        plt.plot(tr['time'], tr['voltage'])
        plt.xlim([1000, 5000])
        plt.ylim([-120, 30])
    plt.tight_layout(pad=2)
    plt.show()
    print(notes[i])
    print('\n\n')
```

```python
# output selected neurons
hof_chk1 = np.load('hof_chk1.npy', allow_pickle=True).tolist()
hof_chk1_keys = sorted(list(hof_chk1.keys()))
hof_chk2 = {}
for i, k in enumerate(hof_chk1_keys):
    if len(notes[i]) == 0:
        hof_chk2[k] = hof_chk1[k]
np.save('hof_chk2.npy', hof_chk2, allow_pickle=True)
```
